## 1. Python Basics

```python
print("Hello, World!")

# Variables
message = "Python is fun!"
print(message)

# Data types
x = 10   # integer
y = 3.14 # float
name = "Aniket" # string
print(type(x), type(y), type(name))
```

## 2. Strings

```python
text = "Hello, Python!"
print(text.lower())
print(text.upper())
print(text.startswith("Hello"))
print(text.find("Python"))
print(text.replace("Python", "World"))
```

## 3. Integers and Floats

```python
a, b = 5, 2
print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a // b)  # Floor division
print(a % b)   # Modulus
print(a ** b)  # Power
```

## 4. Lists, Tuples, Sets

```python
# List
fruits = ["apple", "banana", "cherry"]
fruits.append("orange")
print(fruits)

# Tuple
coordinates = (10, 20)
print(coordinates)

# Set
numbers = {1, 2, 3, 2}
print(numbers)
```

## 5. Dictionaries

```python
student = {"name": "Aniket", "age": 20, "courses": ["Math", "CS"]}
print(student["name"])
print(student.get("age"))
student["phone"] = "123-456"
print(student)
```

## 6. Conditionals and Booleans

```python
x = 10
y = 20
if x < y:
    print("x is less than y")
elif x == y:
    print("x equals y")
else:
    print("x is greater than y")
```

## 7. Loops

```python
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)

count = 0
while count < 3:
    print("Count:", count)
    count += 1
```

## 8. Functions

```python
def greet(name="Guest"):
    return f"Hello, {name}!"

print(greet("Aniket"))
print(greet())

def add(a, b):
    return a + b

print(add(2, 3))
```

## 9. Modules and Standard Library

```python
import math
print(math.pi)
print(math.sqrt(16))

from math import factorial
print(factorial(5))

import os
print(os.getcwd())
```
